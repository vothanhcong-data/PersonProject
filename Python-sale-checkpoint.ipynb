```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
#lọc theo tháng
df['Date']= df['Date'].astype('datetime64[ns]')
df['month_num']= df['Date'].dt.month
df['month_num']

#Quantity by Month,Branch
df_by_period = df.groupby(['month_num','Branch'],as_index = False)['Quantity'].sum()
df_by_period
#add Month_Name
month_name = {}
for i in range(1, 13):
    month_name[str(i)] = f'Tháng {i}'
# Chuyển đổi cột 'month_num' sang 'month_name'
df_by_period['month_num'] = df_by_period['month_num'].astype(str).map(month_name)

# Vẽ biểu đồ lineplot
plt.figure(figsize=(12, 10))
ax = sns.lineplot(data=df_by_period, x='month_num', y='Quantity', hue='Branch')

# Thêm tiêu đề cho biểu đồ
ax.set_title('Quantity by Month and Branch', fontsize=14)

# Thêm nhãn cho trục Y
ax.set_ylabel('Quantity', fontsize=12)
ax.set_xlabel('')
# Tùy chỉnh nhãn trục X
ax.set_xticks(ax.get_xticks())
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, fontsize=10)
# Thêm giá trị 
for line in ax.lines:
    for x, y in zip(line.get_xdata(), line.get_ydata()):
        ax.text(x, y, f'{y:.2f}', fontsize=10)
plt.show()
```

```python
df_quantity_by_gender = df.groupby(['Gender'],as_index = False)['Quantity'].sum()
sns.barplot(df_quantity_by_gender, x="Gender", y="Quantity")

```

```python
# Tổng các Rating dưới 5 
df_rating_low = df[df['Rating'] < 5]
total_rating_byBranch = df_rating_low.groupby(['Branch'], as_index=False)['Rating'].count()
print(total_rating_byBranch)

# tổng số lượng đánh giá
total_ratings = df['Rating'].count()

# tỉ lệ rating dưới 5
rating_percen_low = (df_rating_low['Rating'].count() / total_ratings) * 100

# In kết quả với định dạng phần trăm
print(f"Phần trăm đánh giá dưới 5 là: {rating_percen_low:.2f}%")
```
